Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
x = []
y1 = []
y2 = []
r = 5
def y(x):
  y1 = r* math.sqrt(1 - (x-2)**2/r**2) + 2
  y2 = - r* math.sqrt(1 - (x-2)**2/r**2) + 2
  return y1, y2
for i in range(-3*100, (r+2)*100):
    i /= 100
    x.append(i)
    y1.append(y(i)[0])
    y2.append(y(i)[1]) 
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b


```python
k = [-0.5, 2, 5]
a = [2, -1, 0.5]
b = [1, 4, 0.5]
x = np.linspace(-2*np.pi, 3*np.pi, 201)
for i in range(3):
  plt.plot(x, k[i]*np.cos(x - a[i])+b[i])
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid(True)
plt.show() 
```

exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve
def equations(p):
  x, y = p
  return y - (x**2)+1, np.exp(x) + x*(1-y)-1

x0, y0 = fsolve(equations, (2.5, 5))
x1, y1 = fsolve(equations, (4.5, 20))

print(x0, y0, '\n', x1, y1)

x = np.linspace(-2, 5, 100)
y1 = (x**2) - 1
y2 = np.exp(x)/x
# x !=0 
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")


```
